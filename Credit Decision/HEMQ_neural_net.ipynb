```python
df = pd.read_csv('/content/drive/MyDrive/DSBA 6211/HEMQ.csv')
```

```python
# cleaning steps:
cleaing steps:
1. NA imputation
2. dummy variables
3. scale data
```

```python
# data imputation (decision tree can work with missing values; but this sklearn package cannot work with it)
NA_names =['MORTDUE','VALUE','YOJ','DEROG',
           'DELINQ','CLAGE','NINQ','CLNO','DEBTINC']
df[NA_names]=df[NA_names].fillna(df[NA_names].mean())
df['REASON']=df['REASON'].fillna('DebtCon')
df['JOB']=df['JOB'].fillna('Other')
df.describe(include='all')
```

```python
# generate dummy and combine with the numeric variables together
dummy_data =pd.get_dummies(df[['REASON','JOB']], drop_first=True).astype(int)
print(dummy_data)
```

```python
# scaling
X = pd.concat([df.iloc[:,1:4],df.iloc[:,6:13]],axis=1)

names=X.columns

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(X)
X[names]=scaler.transform(X)

X.describe(include='all')
```

```python
# combine data together
X =pd.concat([X,dummy_data],axis=1)
X.describe()
```

```python
# generate y and data partipation

y= df['BAD']

from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,y,
                                             test_size=0.3,
                                             random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier

NN = MLPClassifier(hidden_layer_sizes=(5,3),max_iter=1000) # first layer 5 and the second layer 3

NN.fit(X_train,y_train)
```

```python
# model evaluation
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot as plt

# confusion matrix
prediction_prob = NN.predict_proba(X_val)
print(NN.score(X_val,y_val)) # accuracy score

# roc curve
nn_auc = roc_auc_score(y_val, prediction_prob[:,1])
print(nn_auc)
```
